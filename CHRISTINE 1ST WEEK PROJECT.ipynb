```python
# Rock Paper Scissors Game
# Human vs Computer
# This program allows a user to play against the computer
# The first to reach majority points wins (2 out of 3 OR 3 out of 5)

import random   # Used to allow the computer to make a random choice

print(" WELCOME TO ROCK PAPER SCISSORS GAME ")

#ask human how many rounds they want (3 or 5)
rounds = int(input("Enter number of rounds:"))

# Majority points needed to win
points_to_win = (rounds // 2) + 1

# Initialize scores
human_score = 0
computer_score = 0


options = ["ROCK", "PAPER", "SCISSORS"]

for round_number in range(1, rounds + 1):

    print(f"\nRound {round_number}")
    print(f"First to {points_to_win} points wins!")


    human = input("Enter ROCK, PAPER or SCISSORS: ")
    computer = random.choice(options)

    # Check if user entered correct word
    if human not in options:
        print("Invalid choice. Try again.")
        continue   # repeat same round

    print("You chose:", human)
    print("Computer chose:", computer)

     # Decide winner
    if human == computer:
        print("It is a tie!")

    elif (human== "rock" and computer == "scissors") or \
         (human == "paper" and computer == "rock") or \
         (human== "scissors" and computer == "paper"):

        print("I win this round!")
        human_score += 1

    else:
        print("Computer wins this round!")
        computer_score += 1

    # Show scores after each round
    print("Score -> You:", human_score, "| Computer:" , computer_score)
     
     
#  Game automatically ends here

print("\nFINAL RESULT" )
if human_score > computer_score:
    print(" I win the game!")
else:
    print(" Computer wins the game!")

```
